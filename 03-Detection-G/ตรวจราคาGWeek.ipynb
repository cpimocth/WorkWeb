```python
thai_months = ["", "ม.ค.", "ก.พ.", "มี.ค.", "เม.ย.", "พ.ค.", "มิ.ย.", "ก.ค.", "ส.ค.", "ก.ย.", "ต.ค.", "พ.ย.", "ธ.ค."]
mytext = f"{thai_months[Month]} {str(Year)[-2:]}"
excel_file_path = current_dir.parent / "01-database" / "ราคากลาง" / f"Concat_ราคากลาง_{mytext}_ชุดรายสัปดาห์.xlsx"
allcenter_clean = pd.read_excel(excel_file_path)

allcenter_clean = allcenter_clean[['Online', 'COMMODITY_CODE', 'ประเภทร้านค้า', 'week1', 'week2', 'week3', 'week4', 'ภาวะ1', 'ภาวะ2', 'ภาวะ3', 'ภาวะ4']]
```

```python
WeekDF = pd.read_excel(Main,engine='calamine')
```

```python
WeekDF['SHOP_NAME']

masterF  = current_dir.parent / "01-database" / "Master_CPI" / "Real_Master_CPI.xlsx"
df_map = pd.read_excel(masterF,sheet_name='shop_manual_cleansing')

def classify_advanced(shop_name):
    if not isinstance(shop_name, str):
        return "ร้านอื่นๆ"
    
    # วนลูปเช็คตามลำดับแถวใน Excel (สำคัญมาก: บนลงล่าง)
    for _, row in df_map.iterrows():
        target = row['Result']
        
        # เตรียม Primary Keywords (แยก comma)
        primaries = [k.strip() for k in str(row['Primary Keywords']).split(',') if k.strip()]
        
        # เตรียม Sub-Keywords (ถ้ามี)
        subs = [k.strip() for k in str(row['Sub Keywords']).split(',') if k.strip() != 'nan']

        # เช็ค Primary: ต้องเจออย่างน้อย 1 คำ
        if any(p in shop_name for p in primaries):
            # ถ้ามี Sub-Keywords: ต้องเจออย่างน้อย 1 คำในนี้ด้วย
            if subs:
                if any(s in shop_name for s in subs):
                    return target
                else:
                    continue # ถ้าไม่เจอ Sub ให้ไปเช็คกฎข้อถัดไป
            
            # ถ้าไม่มี Sub-Keywords แสดงว่าเจอ Primary ก็จบเลย
            return target
            
    return "ร้านอื่นๆ"

WeekDF['ประเภทร้านค้า'] = WeekDF['SHOP_NAME'].apply(classify_advanced)
WeekDF
```

```python
wdf = WeekDF[['MONTH','YEAR','COMMODITY_CODE','DESCRIPTION','ประเภทร้านค้า','WEEK1','WEEK2','WEEK3','WEEK4']].copy()
# wdf = mergde by allcenter_clean

# wdf

# 1. เลือกคอลัมน์ที่ต้องการจาก WeekDF (ตามที่คุณระบุ)
wdf = WeekDF[['MONTH', 'YEAR', 'COMMODITY_CODE', 'DESCRIPTION', 'ประเภทร้านค้า', 'WEEK1', 'WEEK2', 'WEEK3', 'WEEK4']].copy()

# 2. ทำความสะอาด COMMODITY_CODE ให้เป็นประเภทเดียวกัน (เช่น String) เพื่อป้องกัน Error ตอน Merge
wdf['COMMODITY_CODE'] = wdf['COMMODITY_CODE'].astype(str).str.replace(r'\.0$', '', regex=True)
allcenter_clean['COMMODITY_CODE'] = allcenter_clean['COMMODITY_CODE'].astype(str).str.replace(r'\.0$', '', regex=True)



# 3. Merge ข้อมูลเข้าด้วยกัน
# ใช้ how='left' เพื่อยึดข้อมูลใน allcenter_clean เป็นหลัก
# หรือ how='inner' ถ้าต้องการเฉพาะรายการที่มีราคาครบทั้ง 4 Week
merged_wdf = pd.merge( wdf,allcenter_clean, on=['COMMODITY_CODE','ประเภทร้านค้า'], how='left').reset_index(drop=True)
merged_wdf = merged_wdf[merged_wdf['Online']=='Online']
week_cols = ['WEEK1', 'WEEK2', 'WEEK3', 'WEEK4']
merged_wdf['avgWEEK'] = merged_wdf[week_cols].mean(axis=1)

week_cols2 = ['week1', 'week2', 'week3', 'week4']
merged_wdf['avgweek'] = merged_wdf[week_cols2].mean(axis=1)

merged_wdf['diffW1'] = abs(merged_wdf['WEEK1']-merged_wdf['week1'])
merged_wdf['diffW2'] = abs(merged_wdf['WEEK2']-merged_wdf['week2'])
merged_wdf['diffW3'] = abs(merged_wdf['WEEK3']-merged_wdf['week3'])
merged_wdf['diffW4'] = abs(merged_wdf['WEEK4']-merged_wdf['week4'])

merged_wdf = merged_wdf.fillna(0)

merged_wdf['sumdiff'] = merged_wdf['diffW1']+merged_wdf['diffW2']+merged_wdf['diffW3']+merged_wdf['diffW4']

merged_wdf = merged_wdf[merged_wdf['avgweek']!=0]

merged_wdf = merged_wdf[merged_wdf['sumdiff']!=0]


merged_wdf = merged_wdf.reset_index(drop=True)
# แสดงผลลัพธ์
merged_wdf = merged_wdf[['MONTH', 'YEAR', 'COMMODITY_CODE', 'DESCRIPTION', 'ประเภทร้านค้า',
       'WEEK1', 'WEEK2', 'WEEK3', 'WEEK4', 'Online', 'week1', 'week2', 'week3',
       'week4']]
```
