```python
# Parameters
Month = "1"
Year = "2569"
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import os, time, shutil
from datetime import datetime

# # ===== 1. ดึงเดือนและปีปัจจุบันอัตโนมัติ =====
# now = datetime.now()
# Month = str(now.month)     # ได้เลขเดือน เช่น "02"
# Year = str(now.year + 543)          # แปลง ค.ศ. เป็น พ.ศ. เช่น "2569"
session = 'u'                       # ตั้งค่าเริ่มต้นเป็น 'g'

# ===== 1. จัดเตรียมโฟลเดอร์ปลายทาง =====
# สร้างโฟลเดอร์ Data_G_Month ถ้ายังไม่มี
final_dir = os.path.join(os.getcwd(), "Data_U_Week")
if not os.path.exists(final_dir):
    os.makedirs(final_dir)
    print(f"📁 สร้างโฟลเดอร์: {final_dir}")

# โฟลเดอร์สำหรับพักไฟล์ชั่วคราวขณะดาวน์โหลด
download_dir = os.getcwd()

# ===== 2. ตั้งค่า Chrome Options =====
chrome_options = webdriver.ChromeOptions()
prefs = {
    "download.default_directory": download_dir, # พักไว้ที่โฟลเดอร์โค้ดก่อน
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
}
chrome_options.add_experimental_option("prefs", prefs)

# # ===== 3. ลบไฟล์เก่าที่อาจค้างอยู่ใน download_dir =====
# for f in os.listdir(download_dir):
#     if f.startswith(("cpi", "Unconfirmed")) and f.endswith((".xlsx", ".tmp", ".crdownload")):
#         try:
#             os.remove(os.path.join(download_dir, f))
#         except: pass

# ===== 4. เริ่มต้นระบบ Selenium =====
driver = webdriver.Chrome(options=chrome_options)
driver.get("https://sso-tpso.moc.go.th/Login?ReturnUrl=%2F")

# Login
driver.find_element(By.ID, "Username").send_keys("tanaponp")
driver.find_element(By.ID, "Password").send_keys("&5#t2B3&")
driver.find_element(By.ID, "Password").send_keys(Keys.RETURN)
time.sleep(3)

# ไปหน้า Export
driver.get("https://dev-tpso.moc.go.th/export12c/cpi")
time.sleep(3)

# เลือกเดือน/ปี
driver.execute_script(f"document.getElementById('ddlMonth').value = '{Month}';")
driver.execute_script(f"document.getElementById('ddlYear').value = '{Year}';")

# คลิกดาวน์โหลดตาม Session
if session == 'g':
    driver.find_element(By.ID, "btnCpigWeek").click()
    print(f"📂 กำลังดาวน์โหลด CPIG Week: {Month}/{Year} ...")
elif session == 'u':
    driver.find_element(By.ID, "btnCpiuWeek").click()
    print(f"📂 กำลังดาวน์โหลด CPIU Week: {Month}/{Year} ...")

# ===== 5. รอจนโหลดเสร็จ (ใช้ Logic ตรวจสอบไฟล์) =====
print("⏳ รอการดาวน์โหลดเสร็จสิ้น...")
timeout = 120
start_wait = time.time()
downloaded_file = None

while time.time() - start_wait < timeout:
    # ค้นหาไฟล์ .xlsx ที่เพิ่งโหลดมา (ชื่อมักขึ้นต้นด้วย cpig หรือ cpiu)
    files = [f for f in os.listdir(download_dir) if f.endswith(".xlsx") and f.startswith("cpi")]
    if files:
        downloaded_file = files[0]
        # ตรวจสอบว่าไม่มีไฟล์ชั่วคราว (.crdownload) เหลืออยู่
        if not any(f.endswith('.crdownload') for f in os.listdir(download_dir)):
            time.sleep(2) # รอจังหวะเขียนไฟล์ลงดิสก์ให้สนิท
            break
    time.sleep(2)

# ===== 6. ย้ายไฟล์ไปยังโฟลเดอร์ Data_G_Month =====
if downloaded_file:
    old_path = os.path.join(download_dir, downloaded_file)
    new_path = os.path.join(final_dir, downloaded_file)
    
    # ย้ายไฟล์ (ถ้ามีชื่อซ้ำจะเขียนทับ)
    shutil.move(old_path, new_path)
    print(f"✅ ย้ายไฟล์สำเร็จ: {new_path}")
else:
    print("❌ ไม่พบไฟล์ที่ดาวน์โหลดมา")

driver.quit()
```
