```python
import requests
import os

# 1. ตั้งค่าข้อมูลไฟล์
file_id = '1Ars3C5BSAPw1-01z5VJxpcgJbOwJzl0s'
folder_name = 'Master_CPI'  # ชื่อโฟลเดอร์ที่ต้องการสร้าง
file_name = 'Real_Master_CPI.xlsx'

# ลิงก์สำหรับ Download ตรง (ต้องแชร์ไฟล์เป็น "Anyone with the link" ก่อน)
url = f'https://docs.google.com/spreadsheets/d/{file_id}/export?format=xlsx'

# 2. ตรวจสอบและสร้างโฟลเดอร์
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    print(f"📁 สร้างโฟลเดอร์ '{folder_name}' เรียบร้อยแล้ว")
else:
    print(f"✅ พบโฟลเดอร์ '{folder_name}' อยู่แล้ว")

# 3. กำหนดที่อยู่ไฟล์ปลายทาง
save_path = os.path.join(folder_name, file_name)

# 4. เริ่มดาวน์โหลด
print(f"🚀 กำลังดาวน์โหลดไฟล์ไปที่: {save_path} ...")
try:
    response = requests.get(url)
    
    if response.status_code == 200:
        with open(save_path, 'wb') as f:
            f.write(response.content)
        print(f"✨ ดาวน์โหลดสำเร็จ! ไฟล์อยู่ที่: {os.path.abspath(save_path)}")
    else:
        print(f"❌ โหลดไม่ได้ (Status: {response.status_code})")
        print("💡 อย่าลืมตั้งค่าการแชร์ไฟล์ใน Google Drive เป็น 'ทุกคนที่มีลิงก์' (Anyone with the link)")

except Exception as e:
    print(f"❌ เกิดข้อผิดพลาด: {e}")
```
